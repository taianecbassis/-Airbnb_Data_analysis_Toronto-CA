##### **Q3.2. Verificação de outliers via *Boxplot***




Os *Outliers* são medidas que apresentam grande afastamento do restante das observações. A identificação de outliers pode ser realizada por meio da construção do gráfico boxplot. Embora muitas configurações possam ser utilizadas em um boxplot, aquela mais útil para a detecção de outliers é a que emprega a amplitude interquartil e a mediana.

Nesta configuração, a existência destes é detectada a partir da amplitude interquartil (𝐴𝐼𝑄) que corresponde à diferença entre o terceiro e o primeiro quartil, conforme equação abaixo, em que 𝑄3 e 𝑄1 representam o terceiro e o primeiro quartil, respectivamente:

`𝐴𝐼𝑄= 𝑄3 − 𝑄1`




###### *Q3.2.1.1 Boxplot para ```price```*

````python
# @title
# Boxplot - Price
df.price.plot(kind='box', vert=False, figsize=(10,3));
plt.show()

# Quantidade de valores acima de 219 para a variável price
print("\nQuantidade de valores acima de 219: {}".format(df[df.price > 219].shape[0]))
print("Porcentagem de valores acima de 219: {:.4f}%".format(df[df.price > 219].shape[0] / df.shape[0] * 100))
````

###### *Q3.2.1.2 Boxplot para ```minimum_nights```*

###### *Q3.2.1.3 Limpeza dos dados e plotagem dos novos boxplots para as variáveis ```price``` e ```minimum_nights```*

Para tratamento dos dados quanto aos outliers, será utilizada a regra 1,5.𝐴𝐼𝑄. Onde, valores que estejam situados abaixo de 𝑄1 ou acima de 𝑄3 por mais 1,5.𝐴𝐼𝑄 serão considerados outliers sendo excluídos do dataset, segundo equeções abaixo:

- *Limite Superior :* ```𝑄3 + 1,5.𝐴𝐼𝑄```
- *Limite inferior :* ```𝑄1 - 1,5.𝐴𝐼𝑄```

Os valores dos quartis, utilizados para calcular os limites superior e inferior, foram obtidos através do método ```quantile()```.

Analisando os valores dos limites inferiores de ambas as variáveis, `price` e `minimum_nights`, vê-se que esses são negativos. Como nenhuma dessas variáveis pode admitir valores abaixo de zero, é possive concluir que não há outliers negativos no conjunto de dados, porque os valores não podem ser menores que zero. Portanto, entende-se que não há a presença de outliers na cauda inferior da distribuição.

Por fim, quanto a exclusão dos dados, serão retirados do dataset o motante de 18 entradas, correspontes a 0.0864% dos dados, para a variável `price`e, 635 entradas para a variável `minimum_nights`, o que corresponde a 3.0466% do dados desta variável.

Adicionalmente, também foi removida dos dados a coluna da variável `neighbourhood_group`, já que esta está vazia.

````python
# @title
# Calculo dos limites superior e inferior pra as variáveis e definição dos valores de corte
#Quartis - price
q3_price = df.price.quantile(0.75)
print("Q3 - price: {}".format(q3_price))
q1_price = df.price.quantile(0.25)
print("Q1 - price: {}".format(q1_price))

# Limite Superior - price
lim_sup_price = df.price.quantile(0.75) + 1.5 * (df.price.quantile(0.75) - df.price.quantile(0.25))
print("\nLimite Superior - price: {}".format(lim_sup_price))
# Limite inferior - price
lim_inf_price = df.price.quantile(0.25) - 1.5 * (df.price.quantile(0.75) - df.price.quantile(0.25))
print("Limite Inferior - price: {}".format(lim_inf_price))

#Quartis - minimum_nights
q3_minimum_nights = df.minimum_nights.quantile(0.75)
print("\nQ3 - minimum_nights: {}".format(q3_minimum_nights))
q1_minimum_nights = df.minimum_nights.quantile(0.25)
print("Q1 - minimum_nights: {}".format(q1_minimum_nights))

# Limite #Limite Superior - minimum_nights
lim_sup_min_nights = df.minimum_nights.quantile(0.75) + 1.5 * (df.minimum_nights.quantile(0.75) - df.minimum_nights.quantile(0.25))
print("\nLimite Superior - minimum_nights: {}".format(lim_sup_min_nights))
# Limite inferior - minimum_nights
lim_inf_min_nights = df.minimum_nights.quantile(0.25) - 1.5 * (df.minimum_nights.quantile(0.75) - df.minimum_nights.quantile(0.25))
print("Limite Inferior - minimum_nights: {}".format(lim_inf_min_nights))

# Quantidade de valores acima dos Limites Superiores para as variáveis price e minimum_nights
print("\nQuantidade de valores acima do limite superior - price: {}".format(df[df.minimum_nights > lim_sup_price].shape[0]))
print("Porcentagem de valores do limite superior - price: {:.4f}%".format(df[df.minimum_nights > lim_sup_price].shape[0] / df.shape[0] * 100))
print("\nQuantidade de valores do limite superior - minimum_nights: {}".format(df[df.minimum_nights > lim_sup_min_nights].shape[0]))
print("Porcentagem de valores do limite superior - minimum_nights: {:.4f}%".format(df[df.minimum_nights > lim_sup_min_nights].shape[0] / df.shape[0] * 100))
````

````python
# @title
# Exclusão dos outliers do dataset
df_clear = df.copy()
df_clear.drop(df_clear[df_clear.price > lim_sup_price].index, axis=0, inplace=True)
df_clear.drop(df_clear[df_clear.minimum_nights > lim_sup_min_nights].index, axis=0, inplace=True)

# Remoção da coluna neighbourhood_group
df_clear.drop('neighbourhood_group', axis=1, inplace=True)

# Análise da distribuição das variáveis
df_clear.hist(bins=15, figsize=(15,10));

# Checagem do novo shape do dataframe
print("Tamanho do dataset origial:{}".format(df.shape))
print("Tamanho do dataset após limpeza:{}\n".format(df_clear.shape))
````

````python
# @title
# Comparação dos boxplots dos dados originais x dados após limpeza
fig, ax = plt.subplots(2, 2, figsize=(15, 9))

# Boxplot do dataset original - price
sns.boxplot(data=df['price'], ax=ax[0, 0], orient='h', color='#5499c7')
ax[0, 0].set_title("Dataset Original - Boxplot de 'price'", color="#515151", style='italic', fontweight='bold')
ax[0, 0].set_xlabel('price')

# Boxplot dos dados limpos - price
sns.boxplot(data=df_clear['price'], ax=ax[0, 1], orient='h', color='#bb8fce')
ax[0, 1].set_title("Dataset Limpo - Boxplot de 'price'", color="#253760", style='italic', fontweight='bold')
ax[0, 1].set_xlabel('price')

# Boxplot do dataset original - minimum_nights
sns.boxplot(data=df['minimum_nights'], ax=ax[1, 0], orient='h', color='#5499c7')
ax[1, 0].set_title("Dataset Original - Boxplot de 'minimum_nights'", color="#515151", style='italic', fontweight='bold')
ax[1, 0].set_xlabel('minimum_nights')

# Boxplot dos dados limpos - minimum_nights
sns.boxplot(data=df_clear['minimum_nights'], ax=ax[1, 1], orient='h', color='#bb8fce')
ax[1, 1].set_title("Dataset Limpo - Boxplot de 'minimum_nights'", color="#253760", style='italic', fontweight='bold')
ax[1, 1].set_xlabel('minimum_nights')

plt.tight_layout()
plt.show()

````

### **Q4. Análise da correlação entre as variéveis**

A **análise de correlação** é uma técnica estatística usada para medir a força e a direção da relação entre duas ou mais variáveis. O valor da correlação varia entre -1 e 1:

***Correlação positiva:*** Quando uma variável aumenta, a outra também tende a aumentar (próximo de 1).

***Correlação negativa:*** Quando uma variável aumenta, a outra tende a diminuir (próximo de -1).

***Correlação zero: ***Indica que não há relação linear entre as variáveis.

Assim, a matriz de correlação, que é uma tabela que exibe os coeficientes de correlação entre várias variáveis, foi calculada a partir do método `corr()`.

Cada célula na matriz mostra o coeficiente de correlação entre dois pares de variáveis, facilitando a visualização das relações entre múltiplos fatores em um único lugar.

A matriz de correlação em questão foi calculada para as variáveis `price`, `minimum_nights`,`number_of_reviews`, `reviews_per_month`, `calculated_host_listings_count`, e `availability_365`.

Para melhor visualização das correlações encontradas, também foi plotado o *Heatmap *para o conjunto de variáveis mencionados acima.





````python
# @title
# Matriz de correlação do dados
corr = df_clear[['price','minimum_nights','number_of_reviews', 'reviews_per_month',
                'calculated_host_listings_count','availability_365']].corr()

display(corr)
````

````python
# @title
# Heatmap a partir das correlações
sns.heatmap(corr, fmt='.2f', square = True, annot=True);
````

Após a verificação dos valores na matriz de correlação e no Heatmap, foi possível notar que ***não há correlações significativas entre as variáveis***, levando a suposição de que os preços dos imóveis estão relacionados a outros fatores não capturados neste recorte do conjunto de dados.

Assim, é necessário a introdução de outras variáveis do dataset para melhor entendimento do comportamento dos preços de alugéis em Toronto.

### **Q7. Média do mínimo de noites para aluguel (minimum_nights)**

Em Toronto, a média do número mínimo de noites de estadia é de 20 dias.

Para aprofundar essa análise e verificar se o número mínimo de noites varia entre os bairros, agrupamos os imóveis da cidade de Toronto por bairro e analisamos a distribuição do número mínimo de noites de estadia, plotando histogramas separados para cada um.

Com isso, visualizamos como os valores de `minimum_nights` se distribuem em relação à média de 20 noites de estadia, permitindo identificar facilmente os bairros que fogem desse padrão. Os histogramas encontrados permitem observar variações significativas, indicando as áreas que têm um número mínimo de noites de estadia acima ou abaixo da média.


````python
# @title
# Média da coluna `minimum_nights`
df_clear['minimum_nights'].mean()
````

````python
# @title
# Média da coluna `minimum_nights` por bairro
df_clear.groupby(['neighbourhood']).minimum_nights.mean().sort_values(ascending=False)
````

````python
# @title
import matplotlib.pyplot as plt

# Agrupar por bairro
groups = df_clear.groupby('neighbourhood')

# Definir o número de colunas
num_cols = 3
num_neighbourhoods = len(groups)
num_rows = (num_neighbourhoods + num_cols - 1) // num_cols  # Calcula o número de linhas

# Criar a figura e os eixos
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 5 * num_rows))
axes = axes.flatten()  # Transformar em uma lista 1D para facilitar a iteração

# Criar um gráfico separado para cada bairro
for ax, (neighbourhood, group) in zip(axes, groups):
    group['minimum_nights'].hist(bins=30, alpha=0.7, ax=ax)  # Plota o histograma no eixo correspondente
    ax.set_title(f'Distribuição do Número Mínimo de Noites - {neighbourhood}')  # Título
    ax.set_xlabel('Número Mínimo de Noites')  # Rótulo do eixo x
    ax.set_ylabel('Frequência')  # Rótulo do eixo y
    ax.grid(axis='y', alpha=0.75)  # Adiciona uma grade no eixo y

# Remover eixos extras, se houver
for i in range(len(groups), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()  # Exibe os gráficos
````
